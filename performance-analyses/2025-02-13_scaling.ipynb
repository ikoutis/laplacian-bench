```julia
function combine(a, b)
    return [a; b]
end
```

```julia
dic_all = Dict();
for (i, ds) in enumerate(datasets)
    if i == 1
        dic_all = copy(load_object(ds))
    else
        dic = copy(load_object(ds))
        dic_all = merge(combine,dic_all, dic);
    end
end
```
